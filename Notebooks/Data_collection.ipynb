```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 100)

print("Libraries imported successfully ✅")
```

```python

file_path = "../data/raw/inflation.csv"

df = pd.read_csv(file_path)

print("Dataset loaded successfully ✅")
```

```python
print("Dataset Shape:", df.shape)
print("-" * 50)
```

```python
print("Columns:")
print(df.columns)
print("-" * 50)
```

```python
df.head()
```

```python
df.info()
```

```python
missing_values = df.isnull().sum()

missing_values[missing_values > 0].sort_values(ascending=False)
```

```python
df.describe()
```

```python
plt.figure(figsize=(12,6))

if "CPI" in df.columns:
    plt.plot(df.index, df["CPI"])
    plt.title("Headline CPI Trend (Raw)")
    plt.xlabel("Date")
    plt.ylabel("CPI Index")
    plt.grid(True)
    plt.show()
```

```python
plt.figure(figsize=(8,5))

if "Inflation" in df.columns:
    sns.histplot(df["Inflation"], bins=30, kde=True)
    plt.title("Inflation Distribution")
    plt.xlabel("Inflation Rate")
    plt.ylabel("Frequency")
    plt.grid(True)
    plt.show()
```

```python

processed_path = "../data/processed/processed_inflation.csv"

os.makedirs("../data/processed", exist_ok=True)

df.to_csv(processed_path)

print("Processed dataset saved successfully ✅")
```
